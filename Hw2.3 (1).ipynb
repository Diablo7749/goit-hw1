# Hw2.3 — Аналіз датасету Top-50 books

Обробка локального `top50_books.csv`.

```
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('top50_books.csv')
# Перейменування колонок
df.columns = ['name','author','user_rating','reviews','price','year','genre']
print('Перші 5 рядків:')
print(df.head())
print('\nРозмір:', df.shape)

print('\nПропуски по колонках:')
print(df.isna().sum())
print('\nЧи є пропуски?', 'Так' if df.isna().any().any() else 'Ні')

print('\nУнікальні жанри:')
print(df['genre'].unique())

# Гістограма цін
plt.figure()
df['price'].plot(kind='hist')
plt.title('Розподіл цін')
plt.show()

# Статистики по price
print('\nPrice: max, min, mean, median')
print(df['price'].max(), df['price'].min(), df['price'].mean(), df['price'].median())

# Рейтинг
print('\nНайвищий рейтинг у датасеті:', df['user_rating'].max())
print('Кількість книг з таким рейтингом:', (df['user_rating']==df['user_rating'].max()).sum())

# Книга з найбільше відгуків
idx = df['reviews'].idxmax()
print('\nКнига з найбільше відгуків:', df.loc[idx,'name'])

# Найдорожча книга 2015
df2015 = df[df['year']==2015]
if not df2015.empty:
    idx = df2015['price'].idxmax()
    print('\nНайдорожча книга 2015:', df.loc[idx,'name'])

# Скільки Fiction у 2010
cnt = df[(df['genre']=='Fiction') & (df['year']==2010)].shape[0]
print('\nFiction у 2010:', cnt)

# Рейтинг 4.9 у 2010/2011
cnt49 = df[(df['user_rating']==4.9) & (df['year'].isin([2010,2011]))].shape[0]
print('\nКниг з рейтингом 4.9 у 2010/2011:', cnt49)

# Сортування 2015 price<8
subset = df2015[df2015['price']<8].sort_values('price')
if not subset.empty:
    print('\nОстання книга у відсортованому списку (2015, price<8):', subset.iloc[-1]['name'])
else:
    print('\nНема книг 2015 з price<8')

# Агрегація по жанрам
agg = df.groupby('genre')['price'].agg(['min','max'])
print('\nАгрегація min/max по жанрам:')
print(agg)

# Кількість книг на автора
books_per_author = df.groupby('author')['name'].count().to_frame('count')
print('\nРозмір таблиці books_per_author:', books_per_author.shape)
print('Автор з найбільшою кількістю книг:', books_per_author['count'].idxmax(), books_per_author['count'].max())

# Середній рейтинг по автору
avg_rating_author = df.groupby('author')['user_rating'].mean().to_frame('avg_rating')
min_author = avg_rating_author['avg_rating'].idxmin()
print('\nАвтор з мінімальним середнім рейтингом:', min_author, avg_rating_author.loc[min_author,'avg_rating'])

# Concat
joined = pd.concat([books_per_author, avg_rating_author], axis=1)
joined_sorted = joined.sort_values(['count','avg_rating'])
print('\nПерший автор у відсортованому списку:', joined_sorted.index[0])

# Збереження
df.to_csv('hw2_3_processed.csv', index=False)
print('\nЗбережено hw2_3_processed.csv')
```
