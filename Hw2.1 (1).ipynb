# Hw2.1 — Знайомство з Pandas

Зчитування `birth_rate.csv` та виконання всіх кроків завдання.

```
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Читаємо CSV (згенерований локально)
df = pd.read_csv('birth_rate.csv')
print('Перші рядки:')
print(df.head())
print('\nРозмір:', df.shape)

# Заміна '—' (якщо такі є) на NaN
df.replace('—', np.nan, inplace=True)

# Перевірка типів
print('\nТипи до перетворення:')
print(df.dtypes)

# Перетворюємо всі річні колонки на числові
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print('\nТипи після перетворення:')
print(df.dtypes)

# Частка пропусків
print('\nЧастка пропусків по колонках:')
print(df.isnull().sum() / len(df))

# Видаляємо останній рядок (дані по країні)
if 'Україна' in df.iloc[-1,0]:
    df = df.iloc[:-1].reset_index(drop=True)
    print('\nОстанній рядок (агрегат по країні) видалено')

# Заповнюємо NaN середніми значеннями по стовпцях
num_cols = df.select_dtypes(include=[np.number]).columns
for c in num_cols:
    df[c].fillna(df[c].mean(), inplace=True)

# Регіони з народжуваністю у 2019 вище за середній
if '2019' in df.columns:
    avg_2019 = df['2019'].mean()
    regions_higher = df.loc[df['2019'] > avg_2019, 'Region'].tolist()
    print('\nСереднiй по 2019:', avg_2019)
    print('Регіони вищі за середній (2019):', regions_higher)
else:
    print('\nКолонка 2019 відсутня')

# Регіон з найвищою народжуваністю у 2014
if '2014' in df.columns:
    idx = df['2014'].idxmax()
    print('\nРегіон з найвищою народжуваністю у 2014:', df.loc[idx,'Region'])

# Побудова стовпчикової діаграми для 2019
if '2019' in df.columns:
    plt.figure(figsize=(8,4))
    plt.bar(df['Region'], df['2019'])
    plt.xticks(rotation=45)
    plt.title('Народжуваність по регіонах у 2019')
    plt.tight_layout()
    plt.show()

# Збережемо оброблений CSV
df.to_csv('hw2_1_processed.csv', index=False)
print('\nЗбережено hw2_1_processed.csv')
```
